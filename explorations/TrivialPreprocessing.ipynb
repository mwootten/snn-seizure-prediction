```python
# 22 hardcoded because it's a beta: I don't want to deal with the leftover blocks.
def read_chunk(file, chunksize=22):
    return np.fromfile(file, dtype=np.dtype('i4'), count=chunksize)
```

```python
# this is where any refactoring based on a sample-by-sample setup would go
def average_chunks(chunks):
    averages = []
    for i in range(0, len(chunks[0]) - 1):
        averages += [np.mean(chunks[:, i])]
    return averages
```
