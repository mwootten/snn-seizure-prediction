```python
start_scope()

# is this the same τ as τ_mp in Lee et al.?
# the demo uses 10 ms
# Lee et al. is 20 or 200, depending on the data set
# This might be a hyperparameter
tau = 10*ms

eqs = '''
dv/dt = (1-v)/tau : 1
'''

# method: linear -> solve differential equation and plug in
# method: euler  -> use euler's method
G = NeuronGroup(1, eqs, method='euler', threshold='v > 0.8', reset='v=0')
M = StateMonitor(G, 'v', record=True)
run(90*ms)

plot(M.t/ms, M.v[0])
xlabel('Time (ms)')
ylabel('v');
```
